```python
import numpy as np
import matplotlib.pyplot as plt

# Step 2: Prepare the data
colleges = ["Princeton", "Harvey Mudd", "Caltech", "US Naval Academy", "West Point", "MIT", "Lehig University", "NYU-Poly", "Babson", "Stanford"]
mid_career_salary = [137, 135, 127, 122, 120, 118, 118, 117, 117, 114]
tuition = np.array([28540, 40000, 39900, 0, 0, 42050, 43220, 39565, 40400, 54506], dtype=float)  # Ensure 'tuition' is a NumPy array

# Step 3: Compute linear regression coefficients using the closed-form formula
X = np.array([np.ones(len(colleges)), tuition]).T
y = np.array(mid_career_salary)
w = np.linalg.inv(X.T @ X) @ X.T @ y

# Step 4: Create a scatter plot of the data and plot the least squares line
plt.scatter(tuition, mid_career_salary, label="Data")
plt.plot(tuition, w[1] * tuition + w[0], color='red', label="Least Squares Line")
plt.xlabel("Tuition")
plt.ylabel("Mid Career Salary")
plt.title("Scatter Plot and Least Squares Line")
plt.legend()
plt.show()

# Step 5: Compute the determination coefficient (R^2)
y_mean = np.mean(mid_career_salary)
SSR = np.sum((w[1] * tuition + w[0] - y_mean) ** 2)
SST = np.sum((mid_career_salary - y_mean) ** 2)
R2 = SSR / SST

# Step 6: Predict the mid-career salary for tuition = $40,000
tuition_predict = 40000
predicted_salary = w[1] * tuition_predict + w[0]
print(f"Predicted Mid Career Salary for Tuition = $40,000: ${predicted_salary:.2f}")

# Step 7: Remove outliers (zero tuition) and repeat the process
tuition_filtered = tuition[tuition > 0]
mid_career_salary_filtered = [m for i, m in enumerate(mid_career_salary) if tuition[i] > 0]

X_filtered = np.array([np.ones(len(tuition_filtered)), tuition_filtered]).T
y_filtered = np.array(mid_career_salary_filtered)
w_filtered = np.linalg.inv(X_filtered.T @ X_filtered) @ X_filtered.T @ y_filtered
predicted_salary_filtered = w_filtered[1] * tuition_predict + w_filtered[0]
print(predicted_salary_filtered)
plt.scatter(tuition_filtered, mid_career_salary_filtered, label="Data")
plt.plot(tuition_filtered, w_filtered[1] * tuition_filtered + w_filtered[0], color='red', label="Least Squares Line")
plt.xlabel("Tuition")
plt.ylabel("Mid Career Salary")
plt.title("Scatter Plot and Least Squares Line (Outliers Removed)")
plt.legend()
plt.show()

print("R^2 with Outliers:")
print(f"R^2 = {R2:.4f}")

# Step 5: Compute the determination coefficient (R^2) with outliers removed
y_mean_filtered = np.mean(mid_career_salary_filtered)
SSR_filtered = np.sum((w_filtered[1] * tuition_filtered + w_filtered[0] - y_mean_filtered) ** 2)
SST_filtered = np.sum((mid_career_salary_filtered - y_mean_filtered) ** 2)
R2_filtered = SSR_filtered / SST_filtered

print("R^2 without Outliers:")
print(f"R^2 = {R2_filtered:.4f}")
```
